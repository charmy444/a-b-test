```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import mannwhitneyu
from statsmodels.stats.proportion import proportions_ztest
```

## Загрузка данных

```python
df = pd.read_csv("ab_test_results_aggregated_views_clicks_2.csv")

test = df[df["group"] == "test"]
control = df[df["group"] == "control"]

df.info()
```

## EDA + расчет метрики

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

cp1 = sns.countplot(data=test, x="views", ax=axes[0, 0])
cp1.set_xticks(cp1.get_xticks()[::10])
cp1.set_title("test views")

cp2 = sns.countplot(data=test, x="clicks", ax=axes[1, 0])
cp2.set_title("test clicks")

cp3 = sns.countplot(data=control, x="views", ax=axes[0, 1])
cp3.set_xticks(cp3.get_xticks()[::10])
cp3.set_title("control views")

cp4 = sns.countplot(data=control, x="clicks", ax=axes[1, 1])
cp4.set_title("control clicks")

fig.suptitle("Countplots", fontsize=20)
plt.tight_layout()
plt.show()
```

### Добавим метрику CTR

```python
test_ctr = test["clicks"] / test["views"]
control_ctr = control["clicks"] / control["views"]
```

Посмотрим распределение CTR в каждой группе

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(x=test_ctr, label="Test")
sns.kdeplot(x=control_ctr, label="Control")
plt.legend()
plt.show()
```

Проверим распределение с помощью QQ plot

```python
fig, ax = plt.subplots(2, figsize=(5, 10))

stats.probplot(test_ctr, dist="norm", plot=ax[0])
stats.probplot(control_ctr, dist="norm", plot=ax[1])
ax[0].set_title("QQ plot test_ctr")
ax[1].set_title("QQ plot control_ctr")

plt.tight_layout()
plt.show()
```

### Посмотрим как изменилась метрика

```python
mean_test_ctr = test_ctr.mean()
mean_control_ctr = control_ctr.mean()

print(f"Средний CTR в test: {mean_test_ctr}")
print(f"Средний CTR в control: {mean_control_ctr}")
print(f"В тестовой группе средний CTR вырос на {(mean_test_ctr - mean_control_ctr) / mean_control_ctr * 100}%")
```

### Агрегированная CTR

```python
test_agg_ctr = test["clicks"].sum() / test["views"].sum()
control_agg_ctr = control["clicks"].sum() / control["views"].sum()

print(f"Агрегированная CTR в test: {test_agg_ctr}")
print(f"Агрегированная CTR в control: {control_agg_ctr}")
print(f"Агрегированная CTR выросла на {(test_agg_ctr - control_agg_ctr) / control_agg_ctr * 100}%")
```

### Тест пропорций

Так как мы сравниваем клики из просмотров, дополнительно используем тест пропорций для агрегированных CTR.

```python
count = [control["clicks"].sum(), test["clicks"].sum()]
nobs = [control["views"].sum(), test["views"].sum()]

stat, p_value = proportions_ztest(count, nobs)
print(f"Z-statistic: {stat}, p-value: {p_value:.8f}")
```

### Тест Манна-Уитни

Так как пользовательская CTR распределена не нормально, можно дополнительно сравнить группы тестом Манна-Уитни.

```python
stat, p_value = mannwhitneyu(control_ctr, test_ctr, alternative="two-sided")
print(f"U-statistic: {stat}, p-value: {p_value:.8f}")
```

## Вывод

В тестовой группе CTR вырос. Средняя пользовательская CTR выросла примерно на **11.52%**, а агрегированная CTR — примерно на **11.05%**. Оба статистических теста дают маленький p-value, поэтому можно считать, что различие между группами статистически значимо.
